**DATA CLEANING**

```python
# finding missing values
missing_values = df.isnull().sum()
missing_values
```

```python
total_cells = np.product(df.shape)
total_cells
```

```python
total_missing = missing_values.sum()
total_missing
```

```python
precentage_missing = (total_missing/total_cells)*100
precentage_missing
```

```python
#dropped rows
rows_with_na_dropped = df.dropna()
rows_with_na_dropped.head()
#dropped columns
columns_with_na_dropped = df.dropna(axis=1)
columns_with_na_dropped
```

```python
print("Columns in original dataset: %d \n" % df.shape[1])
print("Columns with na's dropped: %d" % columns_with_na_dropped.shape[1])
```

```python
# filling missing values
filling_missing_values = df.loc[:,'entity_id':'region'].head()
filling_missing_values
```

```python
filling_missing_values.fillna(1)
```

```python
filling_missing_values.fillna(method='bfill', axis=1).fillna(1)
```

```python
df.info()
```

```python
df.milestones.round(2)
```

```python
categorical_columns = df.select_dtypes(include = ["object"]).columns.values
df[categorical_columns] = df[categorical_columns].fillna("Unknown")
len(categorical_columns)
```

```python
# numerical data type
numeric_column = df.select_dtypes(exclude = ["object"]).columns.values
df[numeric_column] = df[numeric_column].fillna(1)
```
